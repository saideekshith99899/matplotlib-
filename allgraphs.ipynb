```python
import matplotlib.pyplot as plt
import numpy as np
N=5;
men = [20, 35, 30, 35, 27]
women = [25, 32, 34, 20, 25]
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
x = np.arange(N)
width = 0.35

fig,ax = plt.subplots()
rects1 = ax.bar(x - width/2, men, width)
rects2 = ax.bar(x + width/2, women, width)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy = (rect.get_x() + rect.get_width()/2,height),
                    xytext = (0,3),
                    textcoords = 'offset points',
                    ha = 'center',
                    va = 'bottom');
        
    
autolabel(rects1)
autolabel(rects2)
fig.tight_layout()

plt.show()
```

```python
import cv2
import numpy as np
import os

def pencil_sketch(image_path, output_path):
    # Check if the file exists
    if not os.path.isfile(image_path):
        print(f"Error: The file {image_path} does not exist.")
        return
    
    # Load the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Unable to read the image at {image_path}.")
        return
    
    # Process the image
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    inverted_gray_image = cv2.bitwise_not(gray_image)
    blurred = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)
    inverted_blurred = cv2.bitwise_not(blurred)
    pencil_image = cv2.divide(gray_image, inverted_blurred, scale=256.0)
    cv2.imwrite(output_path, pencil_image)
    cv2.imshow('pencil sketch', pencil_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

input_image_path = 'tree.jpg'
output_image_path = 'tree_sketch.jpg'
pencil_sketch(input_image_path, output_image_path)
```
